```python
# Fitness Recommendation System - Demo
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics.pairwise import cosine_similarity
from IPython.display import display, Markdown, HTML

# Load the saved model and data
print("Loading model and data...")

# Load the model components
with open('fitness_recommendation_model.pkl', 'rb') as f:
    model_components = pickle.load(f)
    
program_transformer = model_components['program_transformer']
encoded_feature_names = model_components['encoded_feature_names']
content_weight = model_components['content_weight']
collab_weight = model_components['collab_weight']

print(f"Model loaded with content_weight={content_weight} and collab_weight={collab_weight}")

# Load the program data
programs_df = pd.read_csv('processed_programs.csv', index_col=0)
program_features_df = pd.read_csv('program_features.csv', index_col=0)

print(f"Loaded {len(programs_df)} fitness programs")
```

```python
# Define helper functions

def content_based_recommendations(user_features, program_features, user_id, top_n=5):
    """Generate program recommendations based on user-program feature similarity"""
    if user_id not in user_features.index:
        print(f"User {user_id} not found")
        return []
        
    # Get the user's feature vector
    user_vector = user_features.loc[user_id].values.reshape(1, -1)
    
    # Calculate cosine similarity to all programs
    similarities = cosine_similarity(user_vector, program_features.values)
    
    # Get the indices of the top N most similar programs
    similar_indices = similarities[0].argsort()[-top_n:][::-1]
    
    # Get the program IDs and similarity scores
    similar_programs = []
    for idx in similar_indices:
        program_id = program_features.index[idx]
        similarity_score = similarities[0][idx]
        similar_programs.append((program_id, similarity_score))
        
    return similar_programs

def find_similar_users(user_features, target_user_id, n_neighbors=5):
    """Find users similar to the target user"""
    if target_user_id not in user_features.index:
        print(f"User {target_user_id} not found")
        return []
    
    # Created a nearest neighbors model - using values to avoid feature name warnings
    knn = NearestNeighbors(n_neighbors=n_neighbors+1, algorithm='auto')
    knn.fit(user_features.values)
    
    # user's feature vector
    user_vector = user_features.loc[target_user_id].values.reshape(1, -1)
    
    # nearest neighbors
    distances, indices = knn.kneighbors(user_vector)
    
    # Get similar users
    similar_user_indices = indices[0][1:]
    similar_distances = distances[0][1:]
    
    # Get user IDs and distances
    similar_users = []
    for idx, dist in zip(similar_user_indices, similar_distances):
        user_id = user_features.index[idx]
        similar_users.append((user_id, dist))
    
    return similar_users

def collaborative_recommendations(user_features, program_features, programs_df, target_user_id, top_n=5):
    """recommendations based on similar users' preferences"""
    # In this demo, we'll simulate collaborative filtering since we don't have actual user data
    # This would normally use find_similar_users, but we'll just use content-based as a fallback
    return content_based_recommendations(user_features, program_features, target_user_id, top_n)

def hybrid_recommendations(user_id, user_features_df, program_features_df, programs_df, 
                          top_n=5, content_weight=0.7, collab_weight=0.3):
    """hybrid recommendations using content-based and collaborative filtering"""
    # content-based recommendations
    content_recs = content_based_recommendations(
        user_features_df, program_features_df, user_id, top_n=top_n*2
    )
    
    # collaborative recommendations
    collab_recs = collaborative_recommendations(
        user_features_df, program_features_df, programs_df, user_id, top_n=top_n*2
    )
    
    # Weight the recommendations
    rec_scores = {}
    
    # content-based recommendations
    for program_id, similarity in content_recs:
        if program_id not in rec_scores:
            rec_scores[program_id] = 0
        # Weight by similarity and content_weight
        rec_scores[program_id] += content_weight * similarity
    
    # collaborative recommendations
    for program_id, score in collab_recs:
        if program_id not in rec_scores:
            rec_scores[program_id] = 0
        # Weight by score and collab_weight
        rec_scores[program_id] += collab_weight * score
    
    # Sort by score
    sorted_recs = sorted(rec_scores.items(), key=lambda x: x[1], reverse=True)
    
    # top N programs
    top_programs = [program_id for program_id, _ in sorted_recs[:top_n]]
    
    # program details
    recommended_programs = programs_df[programs_df['program_id'].isin(top_programs)].copy()
    
    # score to the results
    recommended_programs['match_score'] = recommended_programs['program_id'].map({prog: score for prog, score in sorted_recs})
    
    # match percentage
    max_score = recommended_programs['match_score'].max()
    recommended_programs['match_percentage'] = (recommended_programs['match_score'] / max_score * 100).round().astype(int)
    
    # Sort by score
    recommended_programs = recommended_programs.sort_values('match_score', ascending=False)
    
    return recommended_programs[['program_id', 'title', 'primary_level', 'primary_goal', 'equipment', 
                               'program_length', 'time_per_workout', 'workout_frequency', 'match_percentage']]

def process_user_features(user_df, program_transformer, encoded_feature_names):
    """Process user features to match the program feature space"""
    user_match_features = user_df[['user_id', 'fitness_level', 'goals', 'equipment', 
                                 'preferred_duration', 'preferred_frequency', 'preferred_style']].copy()
    
    user_match_features['primary_goal'] = user_match_features['goals'].apply(
        lambda x: x[0] if isinstance(x, list) and len(x) > 0 else 'General Fitness'
    )

    user_match_features['primary_level'] = user_match_features['fitness_level']
    
    duration_map = {
        '30-45 min': 40,
        '45-60 min': 55,
        '60-75 min': 70,
        '75-90 min': 85,
        '90+ min': 100
    }
    user_match_features['time_per_workout'] = user_match_features['preferred_duration'].map(duration_map)
    
    user_match_features['time_per_workout'] = user_match_features['time_per_workout'].fillna(60)
    
    user_match_features['workout_frequency'] = pd.to_numeric(user_match_features['preferred_frequency'], errors='coerce').fillna(4)
    
    # Map training style
    style_map = {
        'Full Body': 'Full Body',
        'Upper/Lower': 'Upper/Lower',
        'Push/Pull/Legs': 'Push/Pull/Legs',
        'Body Part Split': 'Body Part Split',
        'No preference': 'Other'
    }
    user_match_features['training_style'] = user_match_features['preferred_style'].map(style_map)
    
    user_match_features['training_style'] = user_match_features['training_style'].fillna('Other')
    
    level_to_length = {
        'Beginner': 4,
        'Novice': 8,
        'Intermediate': 12,
        'Advanced': 16
    }
    user_match_features['program_length'] = user_match_features['primary_level'].map(level_to_length)
    user_match_features['program_length'] = user_match_features['program_length'].fillna(8)
    
    def calculate_user_intensity(row):
        intensity = row['workout_frequency'] * 1.0
        
        duration = row['time_per_workout']
        if duration < 45:
            intensity += 0
        elif duration < 60:
            intensity += 1
        elif duration < 90:
            intensity += 2
        else:
            intensity += 3
        
        level_map = {'Beginner': 0, 'Novice': 1, 'Intermediate': 2, 'Advanced': 3}
        intensity += level_map.get(row['primary_level'], 1)
        
        return min(10, intensity)
    
    user_match_features['intensity_score'] = user_match_features.apply(calculate_user_intensity, axis=1)
    
    user_features = user_match_features[['user_id', 'primary_level', 'primary_goal', 'equipment', 
                                       'program_length', 'time_per_workout', 'workout_frequency', 
                                       'training_style', 'intensity_score']]
    
    # same transformer to process user features
    features_to_transform = user_features.drop('user_id', axis=1)
    
    # columns to match expected order
    expected_cols = ['primary_level', 'primary_goal', 'equipment', 'training_style', 
                    'program_length', 'time_per_workout', 'workout_frequency', 'intensity_score']
    
    for col in expected_cols:
        if col not in features_to_transform.columns:
            if col in ['primary_level', 'primary_goal', 'equipment', 'training_style']:
                features_to_transform[col] = 'Other'  
            else:
                features_to_transform[col] = 0  
    
    features_to_transform = features_to_transform[expected_cols]
    
    # Transform
    user_features_encoded = program_transformer.transform(features_to_transform)
    
    if hasattr(user_features_encoded, 'toarray'):
        user_features_array = user_features_encoded.toarray()
    else:
        user_features_array = user_features_encoded
    
    # DataFrame with user_id as index
    user_features_df = pd.DataFrame(
        user_features_array,
        columns=encoded_feature_names,
        index=user_features['user_id']
    )
    
    return user_features_df

def get_program_recommendations(user_profile, content_weight=None, collab_weight=None, num_recommendations=5):
    """Get personalized program recommendations for a user"""
    # optimal weights if not specified
    if content_weight is None or collab_weight is None:
        content_weight = model_components['content_weight']
        collab_weight = model_components['collab_weight']
    
    # temporary user ID
    user_id = 'temp_user'
    user_profile['user_id'] = user_id
    
    # temporary dataframe with the new user
    temp_user_df = pd.DataFrame([user_profile])
    
    # user features
    temp_user_features_df = process_user_features(temp_user_df, program_transformer, encoded_feature_names)
    
    # recommendations
    recommendations = hybrid_recommendations(
        user_id, temp_user_features_df, program_features_df, programs_df,
        top_n=num_recommendations, content_weight=content_weight, collab_weight=collab_weight
    )
    
    return recommendations
```

```python
# interactive UI for user input and recommendations
from ipywidgets import widgets
from IPython.display import display, clear_output

def create_user_profile_ui():
    """Create an interactive UI for collecting user information"""
    style = {'description_width': '25%'}
    layout = widgets.Layout(width='70%')
    
    # widgets
    fitness_level = widgets.Dropdown(
        options=['Beginner', 'Novice', 'Intermediate', 'Advanced'],
        value='Novice',
        description='Fitness Level:',
        style=style,
        layout=layout
    )
    
    goals = widgets.SelectMultiple(
        options=['General Fitness', 'Weight Loss', 'Strength', 'Hypertrophy', 
                 'Bodybuilding', 'Powerlifting', 'Athletics', 'Endurance',
                 'Muscle & Sculpting', 'Bodyweight Fitness'],
        value=['General Fitness'],
        description='Goals (select multiple):',
        style=style,
        layout=layout
    )
    
    equipment = widgets.Dropdown(
        options=['At Home', 'Dumbbell Only', 'Full Gym', 'Garage Gym'],
        value='At Home',
        description='Equipment:',
        style=style,
        layout=layout
    )
    
    preferred_duration = widgets.Dropdown(
        options=['30-45 min', '45-60 min', '60-75 min', '75-90 min', '90+ min'],
        value='45-60 min',
        description='Workout Duration:',
        style=style,
        layout=layout
    )
    
    preferred_frequency = widgets.Dropdown(
        options=[('3 days/week', 3), ('4 days/week', 4), ('5 days/week', 5), ('6 days/week', 6)],
        value=4,
        description='Workout Frequency:',
        style=style,
        layout=layout
    )
    
    preferred_style = widgets.Dropdown(
        options=['Full Body', 'Upper/Lower', 'Push/Pull/Legs', 'Body Part Split', 'No preference'],
        value='Full Body',
        description='Training Style:',
        style=style,
        layout=layout
    )
    
    # submit button
    submit_button = widgets.Button(
        description='Generate Recommendations',
        button_style='primary',
        tooltip='Generate workout program recommendations',
        icon='check'
    )
    
    output = widgets.Output()
    
    # display widgets
    display(fitness_level, goals, equipment, preferred_duration, 
            preferred_frequency, preferred_style, submit_button, output)
    
    def on_button_click(b):
        with output:
            clear_output()
            
            user_profile = {
                'fitness_level': fitness_level.value,
                'goals': list(goals.value),
                'equipment': equipment.value,
                'preferred_duration': preferred_duration.value,
                'preferred_frequency': preferred_frequency.value,
                'preferred_style': preferred_style.value,
            }
            
            # display user profile
            print("User Profile:")
            for key, value in user_profile.items():
                print(f"- {key}: {value}")
            
            # recommendations
            recommendations = get_program_recommendations(user_profile, num_recommendations=5)
            
            # display recommendations
            print("\nRecommended Fitness Programs:")
            for i, row in enumerate(recommendations.itertuples(), 1):
                display(Markdown(f"### {i}. {row.title} (Match: {row.match_percentage}%)"))
                display(Markdown(f"- **Level**: {row.primary_level}"))
                display(Markdown(f"- **Goal**: {row.primary_goal}"))
                display(Markdown(f"- **Equipment**: {row.equipment}"))
                display(Markdown(f"- **Program Length**: {row.program_length} weeks"))
                display(Markdown(f"- **Workout Duration**: {row.time_per_workout} min/workout"))
                display(Markdown(f"- **Frequency**: {row.workout_frequency} workouts/week"))
                display(HTML("<hr style='border-top: 1px solid #ddd;'>"))
            
    submit_button.on_click(on_button_click)
```

```python
# predefined user profiles to demonstrate the model

def simulate_user_recommendations():
    """Simulate recommendations for different user profiles to demonstrate the model"""
    # predefined user profiles
    user_profiles = [
        {
            'name': 'Beginner at Home',
            'fitness_level': 'Beginner',
            'goals': ['General Fitness'],
            'equipment': 'At Home',
            'preferred_duration': '30-45 min',
            'preferred_frequency': 3,
            'preferred_style': 'Full Body'
        },
        {
            'name': 'Intermediate Bodybuilder',
            'fitness_level': 'Intermediate',
            'goals': ['Bodybuilding', 'Hypertrophy'],
            'equipment': 'Full Gym',
            'preferred_duration': '75-90 min',
            'preferred_frequency': 5,
            'preferred_style': 'Push/Pull/Legs'
        },
        {
            'name': 'Advanced Powerlifter',
            'fitness_level': 'Advanced',
            'goals': ['Strength', 'Powerlifting'],
            'equipment': 'Full Gym',
            'preferred_duration': '90+ min',
            'preferred_frequency': 4,
            'preferred_style': 'Upper/Lower'
        }
    ]
    
    # recommendations for each profile
    for profile in user_profiles:
        print(f"User Profile: {profile['name']}")
        print(f"- Fitness Level: {profile['fitness_level']}")
        print(f"- Goals: {profile['goals']}")
        print(f"- Equipment: {profile['equipment']}")
        print(f"- Preferred Duration: {profile['preferred_duration']}")
        print(f"- Preferred Frequency: {profile['preferred_frequency']} days/week")
        print(f"- Preferred Style: {profile['preferred_style']}")
        
        # recommendations
        recommendations = get_program_recommendations(profile, num_recommendations=3)
        
        # recommendations
        print("\nRecommended Fitness Programs:")
        for i, row in enumerate(recommendations.itertuples(), 1):
            print(f"{i}. {row.title} (Match: {row.match_percentage}%)")
            print(f"   - Level: {row.primary_level}")
            print(f"   - Goal: {row.primary_goal}")
            print(f"   - Equipment: {row.equipment}")
            print(f"   - Program Length: {row.program_length} weeks")
            print(f"   - Workout Duration: {row.time_per_workout} min/workout")
            print(f"   - Frequency: {row.workout_frequency} workouts/week")
        
        print("\n" + "-"*80 + "\n")
```

```python
# Run the interactive UI
print("Interactive User Profile - Generate your own recommendations")
print("="*60)
create_user_profile_ui()
```
